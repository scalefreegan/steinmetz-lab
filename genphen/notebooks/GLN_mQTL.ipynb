# Glutamine mQTL

Candidate list for glutamine mQTL

GLN mQTLs required a higher than normal FDR: 10%. Like Phe and some of the other amino acids, it was difficult to find a convincing candidate based on available data. I would be uncertain about this prediction, but growing evidence suggests widespread effects in genes functioning in ubiquitination

**SNT2** was chosen because of its functional relationship to E3 ubiquitin ligases, which seems to be a common thread tying together many of the mQTL predictions...It has a number of SNPs with moderate effect and is not too far from the peak of the mQTL interval.

# SNT2

From SGD:

Subunit of Snt2C complex, RING finger ubiquitin ligase (E3); physically associates with Ecm5p and Rpd3p; along with Ecm5p, recruits Rpd3p to small number of promoters; colocalizes with Ecm5p, independently of Rpd3p, to promoters of stress response genes upon oxidative stress; involved in ubiquitylation, degradation of excess histones; interacts with Ubc4p; role in regulating genes encoding amine transporters; relocalizes from nucleus to cytoplasm upon DNA replication stress

## Metabolite levels conditioning on SNT2 SNP @ chrVII:262283..262283

SNT2 SNP @ chrVII:262283..262283 causes missense mutation in SNT2 (YGL131C): N1193D

Mutation is in the PHD-zinc-finger like domain

SGD variant viewer: http://www.yeastgenome.org/variant-viewer#/S000003099?_k=y68zv6

```r
# find marker
gr <- GRanges(seqnames = Rle(c("chr07")), ranges = IRanges(262283,262283))
thismrk = names(mrk[which(mrk == gr),])

# add geno column to endometabolite data
genotype = geno[thismrk,]
genotype[genotype==1] = "S"
genotype[genotype==2] = "Y"
genotype = data.frame(strain = names(genotype), geno = genotype)

endometabolite2plot = merge(endometabolite,genotype,by = "strain")

# remove strains w/o genotype at this marker
if (length(which(is.na(endometabolite2plot$geno)))>0) {
    data2plot = endometabolite2plot[-which(is.na(endometabolite2plot$geno)),]
} else {
    data2plot = endometabolite2plot
}
# remove NA values
if (length(which(is.na(data2plot$value.log2)))>0) {
    data2plot = data2plot[-which(is.na(data2plot$value.log2)),]
}
data2plot$geno = as.factor(data2plot$geno)

# calculate mean levels per metabolite
mo = data2plot %>% group_by(metabolite) %>% summarise(m_mean = mean(value.log2, na.rm = T))
mo = mo[order(mo$m_mean, decreasing = F),]

data2plot$metabolite = factor(data2plot$metabolite, levels = mo$metabolite)

# average replicates
data2plot = data2plot %>% group_by(metabolite, time, strain) %>% do({
        data.frame(value = mean(.$value.log2, na.rm =T), geno = .$geno[1])
 })
```

```r
# plot
options(repr.plot.height = 5)
options(repr.plot.width = 8)
p <- ggplot(data2plot, aes(geno, value)) + 
    geom_boxplot(aes(fill = geno), show.legend = T) +
    facet_wrap(~ metabolite, scales = "free_y") + 
    labs(title = paste("Metabolite levels, all timepoints conditioned on", thismrk), 
         y = "log2[X]", x = "Genotype") 
p
```

```r
options(repr.plot.width = 8)
options(repr.plot.height = 2)
p <- ggplot(data2plot %>% filter(metabolite %in% "GLN"), aes(factor(time), value)) + 
    geom_boxplot(aes(fill = geno), show.legend = T) +
    labs(title = paste("GLN per timepoint conditioned on", thismrk), 
         y = "log2[X]", x = "Relative Time (hr)") 
p
```

```r
Mmrk_dif_sig = data2plot %>% filter(metabolite %in% "GLN") %>% group_by(metabolite, time) %>% do({
    data = .
    sparent = data[data$geno=="S","value"]
    yparent = data[data$geno=="Y","value"]
    test = t.test(sparent$value, yparent$value)
    pval = test$p.value
    return(data.frame(pval))
})
Mmrk_dif_sig$pval = p.adjust(Mmrk_dif_sig$pval,method="BH")
Mmrk_dif_sig$mlog10.pval = -log10(Mmrk_dif_sig$pval)
#Mmrk_dif_sig = acast(Mmrk_dif_sig, metabolite ~ time, value.var = "pval")

Mmrk_dif_sig
```
