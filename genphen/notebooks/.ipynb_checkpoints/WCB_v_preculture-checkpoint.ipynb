```r
endo_raw = read.delim("/g/steinmetz/brooks/genphen/metabolome/validation/endometabolome_classic_WCB.txt")
endo_mc = melt(endo_raw, id.vars = c("TYPE","TIME"), variable.name = "metabolite")
endo_mc$value = as.numeric(endo_mc$value)
endo_mc[which(endo_mc$value <= 0), "value"] = NA
# shift for log
endo_mc$value = endo_mc$value + 1
endo_mc$value.log2 = log2(endo_mc$value)

method = grepl("star$", endo_mc$TYPE)
method[method==F] = "A/B/C"
method[method==T] = "A*/B*/C*"
endo_mc$method = factor(method)
# exo = read.delim("/g/steinmetz/brooks/genphen/metabolome/validation/exometabolome_classic_WCB.txt")
```

## Can one distinguish WCB from pre-culture method?

Overall reproducibility

```r
m1r1 = endo_mc %>% filter(method=="A/B/C", TYPE == "A") %>% select(value.log2)
m1r2 = endo_mc %>% filter(method=="A/B/C", TYPE == "B") %>% select(value.log2)
m1r3 = endo_mc %>% filter(method=="A/B/C", TYPE == "C") %>% select(value.log2)
m2r1 = endo_mc %>% filter(method=="A*/B*/C*", TYPE == "Astar") %>% select(value.log2)
m2r2 = endo_mc %>% filter(method=="A*/B*/C*", TYPE == "Bstar") %>% select(value.log2)
m2r3 = endo_mc %>% filter(method=="A*/B*/C*", TYPE == "Cstar") %>% select(value.log2)
all_pairs = as.matrix(data.frame(m1r1, m1r2, m1r3, m2r1, m2r2, m2r3))
colnames(all_pairs) = c("A","B","C","A*","B*","C*")
```

```r
options(repr.plot.width = 8)
options(repr.plot.height = 8)
heatpairs(as.matrix(all_pairs),main = "Comparison of metabolite concetrations: Classic v WCB, log2[uM]")
```

```r
allReps = endo_mc %>% group_by(metabolite, TIME, method) %>% do({
    thismethod = levels(.$method[1])[.$method[1]]
    if (thismethod == "A/B/C") {
        ref = "A"
    } else {
        ref = "Astar"
    }
    #val.x = . %>% filter(TYPE == ref)
    val.x = filter(., TYPE == ref)
    val.x = val.x$value.log2
    val.y = filter(., TYPE != ref)
    val.y = val.y$value.log2
    o = cbind(val.x,val.y)
    if (dim(o)[1] == 2 && dim(o)[2] == 2) {
        return(data.frame(o))
    } else {
        return(data.frame())
    }
    
})
```

### Comparison of concentrations across methods and replicates

```r
options(repr.plot.width = 5)
options(repr.plot.height = 4)
allReps$val.z = allReps$val.y
# Add noise component to data
allReps$val.z[allReps$method=="A/B/C"] =  allReps$val.z[allReps$method=="A/B/C"] + 
    rnorm(length( allReps$val.z[allReps$method=="A/B/C"]),0,1)
p <- ggplot(allReps, aes(x = val.x, y = val.y)) + 
    geom_point(aes(color = method)) + 
    facet_wrap(~ method) +
    xlab("Replicate 1, log2[uM]") +
    ylab("Replicate 2/3, log2[uM]")
p
```

### What one would expect if A\*, B\* ,C\* had higher precision

```r
p <- ggplot(allReps, aes(x = val.x, y = val.z)) + 
    geom_point(aes(color = method)) + 
    facet_wrap(~ method) +
    xlab("Replicate 1, log2[uM]") +
    ylab("Replicate 2/3, log2[uM]")
p
```

```r
options(repr.plot.width = 10)
options(repr.plot.height = 8)
p <- ggplot(endo_mc, aes(factor(metabolite), value.log2))
p + geom_boxplot(aes(fill=factor(method))) + facet_wrap(~ TIME, ncol = 1)
```

```r
method_cor = endo_mc %>% group_by(metabolite) %>% do({
    this1 = .[.$method=="A/B/C",] %>% group_by(TIME) %>% do({
        data.frame(mmean = mean(.$value.log2, na.rm = T))
    })
    this2 = .[.$method=="A*/B*/C*",] %>% group_by(TIME) %>% do({
        data.frame(mmean = mean(.$value.log2, na.rm = T))
    })
    data.frame(
        cor_m1_m2 = cor(this1$mmean, this2$mmean, use = "pair")
    )
})
```

```r
options(repr.plot.width = 10)
options(repr.plot.height = 3)
hist(method_cor$cor_m1_m2, breaks = 10, xlab = "Correlation, Pre-culture v. WCB", 
     main = "Similarity of metabolite dynamics in Pre-culture v. WCB", xlim = c(-1,1))
```

```r
options(repr.plot.width = 10)
options(repr.plot.height = 10)
p <- ggplot(endo_mc %>% filter(value.log2 > 0), aes(x = TIME, y = value.log2, color = method)) +
    geom_point() +
    geom_smooth(method = "loess") + 
    facet_wrap(~ metabolite, scales = "free_y")
    #facet_wrap(~ metabolite)
p
```

```r
endo_mc_test = endo_mc %>% filter(value.log2 > 0)

levene.test(endo_mc_test$value.log2, endo_mc_test$method, location="median", correction.method="zero.correction")
```

From adivce of Lin Gen and Chenchen: look at cv per timepoint, mean cv?, dispersion

CV per timepoint

```r
endo_cv = endo_mc_test %>% group_by(metabolite, TIME, method) %>% do({
    thissd = sd(.$value.log2, na.rm = T)
    thismean = mean(.$value.log2, na.rm = T)
    return(data.frame(cv = thissd/thismean))
})
```

```r
cv_m1 = endo_cv %>% filter(method == "A/B/C")
cv_m2 = endo_cv %>% filter(method == "A*/B*/C*")
mcv = merge(cv_m1, cv_m2, by = c("metabolite", "TIME"))
options(repr.plot.width = 7)
options(repr.plot.height = 4)
p <- ggplot(mcv, aes(x = cv.x, y = cv.y)) + 
    geom_point(aes(color = metabolite)) +
    xlim(c(0,1)) +
    ylim(c(0,1)) +
    xlab("CV, A/B/C") +
    ylab("CV, A*/B*/C*") + 
    geom_abline(intercept = 0, slope = 1)
p
```
