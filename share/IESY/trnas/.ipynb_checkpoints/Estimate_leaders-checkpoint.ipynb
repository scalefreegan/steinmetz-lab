# Estimate length of leader sequences from the ARMseq alignments

The default leader/trailer sequence length from [GtRNAdb](http://gtrnadb2009.ucsc.edu/Scere3/Scere3-by-locus.html) is 10nt for both the leader and trailer. There has been some thought that this is an arbitrary definition.

Here I look into the ARMSeq data to see whether there is experimental support for 10nt cutoff.

````r
thesenames = c("seqid","source","type","start","end","width","strand","empty","name","pos","count")
untreated.coverage = read_tsv("/g/steinmetz/project/IESY/trna/results/ARMseq-test/ARM-Seq/untreated_coverage.txt",
                            col_names = thesenames) %>% mutate(name = gsub("ID=","",name)) %>%
                            mutate(width = width(IRanges(start,end))) %>% 
                            group_by(name) %>%
                            mutate(rev.pos = rev(pos), alkb = F)
treated.coverage = read_tsv("/g/steinmetz/project/IESY/trna/results/ARMseq-test/ARM-Seq/treated_coverage.txt",
                            col_names = thesenames) %>% mutate(name = gsub("ID=","",name)) %>%
                            mutate(width = width(IRanges(start,end))) %>% 
                            group_by(name) %>%
                            mutate(rev.pos = rev(pos), alkb = T)
thisdata = bind_rows(untreated.coverage, treated.coverage)
````

````r
thisdata = filter(thisdata,pos<=100 | rev.pos <= 100) %>% 
    group_by(alkb,name) %>% mutate(abs.pos = ifelse(strand=="+",order(pos),order(rev.pos)), 
                              loc = ifelse(abs.pos<=100,"upstream","downstream"))
````

# Plot 1: Coverage in leader/trailer regions of tRNA genes

Here I summarize all sequencing reads 100 nt upstream and downstream of tRNA genes to assess whether there is support for the 10nt leader/trailer cutoff

````r
options(repr.plot.width = 8)
options(repr.plot.height = 3)
p1 <- ggplot(thisdata) + geom_bar(aes(x = abs.pos-100, y=count, fill = alkb),stat="identity") + 
    geom_vline(xintercept = -10, size=0.25) + geom_vline(xintercept = 10, size=0.25) +
    geom_vline(xintercept = 0, size=1, color = "red") + xlab("Position relative to tRNA gene body") +
    annotate("text", label = "Upstream", x = -75, y = 20000, size = 4) + 
    annotate("text", label = "Downstream", x = 75, y = 20000, size = 4)
p1
````

# Figure 2: Fraction of counts covered at particular position 

````r
thisdata2 = thisdata %>% group_by(alkb,loc,abs.pos) %>% summarise(sum.counts = sum(count)) %>% 
    ungroup() %>% 
    mutate(abs.pos = ifelse(loc=="upstream",rev(abs.pos)-100,abs.pos-100)) %>%
    group_by(alkb,loc) %>% 
    arrange(abs.pos) %>% 
    mutate(cum.sum = cumsum(sum.counts)/sum(sum.counts))
thisdata2$loc = fct_rev(thisdata2$loc)
````

````r
p2 <- ggplot(thisdata2, aes(x = abs.pos, y = cum.sum, color = alkb)) + geom_step(size=1) +
    facet_wrap(~ loc) + 
    geom_vline(xintercept = 10, size=0.25)
p2
````

# Results

10nt seems to be a reasonable cutoff for defining leader/trailer sequences given ARMseq data. This definition roughly hits the "elbow" of the ecdf distribution. Lengthening the defintion of leader/trailer would introduce ambiguities in the Salmon quantification (i.e. make it think that transcript should be more differentiable than they actually are). There is not sufficient evidence to warrant changing this defintion at the moment given this tradeoff.
