```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from importlib.machinery import SourceFileLoader
import sys, os
import site
# add conda packages path
sys.path.append("/g/steinmetz/brooks/anaconda/envs/crimp/lib/python3.5/site-packages/")
site.addsitedir("/g/steinmetz/brooks/anaconda/envs/crimp/lib/python3.5/site-packages/")
import pickle
import pandas as pd
import time
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import random
import networkx as nx
import itertools
import glob

sys.path.append("/g/steinmetz/project/IESY/software/crimp/")
from crimp import utils
```

```python
marvelDIR="/g/steinmetz/project/IESY/sequencing/Results/PacBio/marvel/"
files1 = glob.glob(marvelDIR + "filtered/*.fasta")
files2 = glob.glob(marvelDIR + "filtered_m/*.fasta")
files3 = glob.glob(marvelDIR + "filtered_mm/*.fasta")
```

```python
def keepLongest(i, name):
    reads = utils.readFASTA_SeqIO(i)
    longest = 0
    for read in reads.values():
        #print(len(read))
        if len(read) > longest:
            o = read
            longest = len(read.seq)
    o.id = name
    o.name = name
    return(o)

# keep only longest contig
seqs1 = [keepLongest(i, "contig" + i.rstrip(".fasta").split("_")[-1]) for i in files1]
seqs2 = [keepLongest(i, "contig" + i.rstrip(".fasta").split("_")[-1]) for i in files2]
seqs3 = [keepLongest(i, "contig" + i.rstrip(".fasta").split("_")[-1]) for i in files3]

try:
    os.mkdir(marvelDIR + 'filtered/combined/')
    os.mkdir(marvelDIR + 'filtered_m/combined/')
    os.mkdir(marvelDIR + 'filtered_mm/combined/')
except:
    pass

# write contigs
SeqIO.write(seqs1, marvelDIR + 'filtered/combined/combined_longest.fasta', "fasta")
SeqIO.write(seqs2, marvelDIR + 'filtered_m/combined/combined_longest.fasta', "fasta")
SeqIO.write(seqs3, marvelDIR + 'filtered_mm/combined/combined_longest.fasta', "fasta")
```
